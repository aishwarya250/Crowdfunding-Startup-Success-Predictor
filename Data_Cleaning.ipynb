Ideally,missing values in avg_time_bw_rounds_float with avg time bw first and last round

but since first and last funding dates were same dates for all missing values, diff = 0 or time_bw_first_rounds = 0 and division would be throw exception. Also logically, we could just say avg_time_bw_rounds = 0 

```python

a= clean_companies_usa_df['avg_time_bw_rounds_float'].isna()
b= clean_companies_usa_df['last_funding_at']==clean_companies_usa_df['first_funding_at']

temp=pd.DataFrame(np.logical_and(a,b),columns=['val'])

temp.val.value_counts()
```
